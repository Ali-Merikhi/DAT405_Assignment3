```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Reads the csv file
df = pd.read_csv('assignment3-data-1.csv')
k = 4

# Variable for the values in columns: phi and psi
vals1 = df.loc[:, ['phi', 'psi']].values

# Sets the parameters for the model
kmeans_model = KMeans(n_clusters=k, init='k-means++', max_iter=400, random_state=42)
kmeans_model.fit(vals1)
labels = kmeans_model.labels_

# Gets the centroids for every cluster
centers = kmeans_model.cluster_centers_
y_kmeans = kmeans_model.fit_predict(vals1)

# Sets the figure size
plt.figure(figsize=(20, 10))

# Plots the clusters
plt.scatter(vals1[y_kmeans == 0, 0], vals1[y_kmeans == 0, 1], s=100, c='green', label='Cluster 1')
plt.scatter(vals1[y_kmeans == 1, 0], vals1[y_kmeans == 1, 1], s=100, c='red', label='Cluster 2')
plt.scatter(vals1[y_kmeans == 2, 0], vals1[y_kmeans == 2, 1], s=100, c='blue', label='Cluster 3')
plt.scatter(vals1[y_kmeans == 3, 0], vals1[y_kmeans == 3, 1], s=100, c='orange', label='Cluster 4')

# Plots the centroids as black dots
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=75)

# Text titles for the figure and the axes
plt.title('Combinations of Psi and Phi in proteins')
plt.xlabel('phi')
plt.ylabel('psi')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Reads the csv file
df = pd.read_csv('assignment3-data-1.csv')

# Variable for the values in columns: phi and psi
vals2 = df.loc[:, ['phi', 'psi']].values

# array for the inertia
inert = []

for n in range(1, 6):
    # Sets the parameters for the k-means model
    kmeans_model = KMeans(n_clusters=n, init='k-means++', max_iter=500, random_state=42)
    kmeans_model.fit(vals2)

    # Adds the inertia of the model to the array
    inert.append(kmeans_model.inertia_)

plt.figure(1, figsize=(20, 10))

# Creates the dots in the plot
plt.plot(np.arange(1, 6), inert, 'o')

# Draws the lines between the dots
plt.plot(np.arange(1, 6), inert, '-', alpha=0.5)

# Sets labels for the axes
plt.xlabel('Amount of clusters')
plt.ylabel('Inertia')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd

# n = Number of removed points
for n in range(10000, 25001, 7500):
    # Reads the csv file
    df = pd.read_csv('assignment3-data-1.csv')
    df_remove = df

    # Gives randomly chosen points
    remove_rand = np.random.choice(df_remove.index, n, replace=False)

    # Removes the randomly chosen points
    df_rand_remove = df_remove.drop(remove_rand)

    # k = 3, since 3 is what we want to validate
    x = df_rand_remove.iloc[:, 3:5]
    k_means = KMeans(3)
    k_means.fit(x)

    # Makes predictions on the clusters
    clusters = k_means.fit_predict(x)

    # Plots the grapg
    data = df_rand_remove.copy()
    data['clusters'] = clusters

    # Made the size a bit smaller this time, so we can really see the removed points
    plt.scatter(data['phi'], data['psi'], c=data['clusters'], cmap='rainbow', s=2)
    plt.ylabel('psi')
    plt.xlabel('phi')
    plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

# Reads the csv-file
df = pd.read_csv('assignment3-data-1.csv')

# Gets the columns
x = df['phi']
y = df['psi']

column_stack = np.column_stack((x, y))
x_arr = []

# Rotating by 360
for i in column_stack:
    (x_axis, y_axis) = i
    if y_axis < -100:
        y_axis += 360
    if x_axis < 0:
        x_axis += 360
    x_arr.append([x_axis, y_axis])

# Initializes K-means algorithm with 3 clusters
k_means = KMeans(n_clusters=3, random_state=0, n_init=10).fit(x_arr)

y_k_means = k_means.predict(x_arr)
x_arr = np.array(x_arr)

# Creates the scatter plots
plt.scatter(x_arr[:, 0], x_arr[:, 1], c=y_k_means, s=1, cmap='plasma')
centers = k_means.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=50)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import DBSCAN, KMeans
import numpy as np
from sklearn.preprocessing import StandardScaler

# Reads the data from the csv-file
df = pd.read_csv("assignment3-data-1.csv")  # importing the dataset

# Gets the columns
x = df['phi']
y = df['psi']

column_stack = np.column_stack((x, y))
x_arr = []

# Rotating by 360
for i in column_stack:
    (x_axis, y_axis) = i
    if y_axis < -100:
        y_axis += 360
    if x_axis < 0:
        x_axis += 360
    x_arr.append([x_axis, y_axis])

x_arr = np.array(x_arr)

# Creates a copy of the data set
data = df.copy()

for i, row in data.iterrows():
    phi = x_arr[i][0]
    psi = x_arr[i][1]
    data.at[i, 'phi'] = phi
    data.at[i, 'psi'] = psi

# Initializes the DBSCAN
db = DBSCAN(min_samples=100, eps=10).fit(df[['phi', 'psi']])

# Sets the labels to its residue name
data["label"] = db.labels_
outlier_arr = data[data.label == -1]

# Gets the amount of outliers
outliers_amount = outlier_arr["residue name"].value_counts().to_frame()

print(outliers_amount)

# Plots the bar chart
outliers_amount.plot.bar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import DBSCAN, KMeans
import numpy as np

# Reads the data from the csv-file
df = pd.read_csv("assignment3-data-1.csv")  # importing the dataset

pro_res = df[df['residue name'] == 'PRO']
pro_phi = pro_res['phi']
pro_psi = pro_res['psi']

pro = np.column_stack((pro_phi, pro_psi))

pro_size = len(pro_res) / len(df)
print(pro_size * 100, '%')

db = DBSCAN(eps=10, min_samples=6)
y_prediction = db.fit_predict(pro)
db.fit(pro)

# Creates the scatter plot
plt.scatter(pro[:, 0], pro[:, 1], c=y_prediction, cmap='plasma', s=10)
plt.title('Stratification')
plt.xlabel('phi')
plt.ylabel('psi')
plt.show()
```
